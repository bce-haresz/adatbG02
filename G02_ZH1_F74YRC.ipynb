1\. Feladat

```sql
--- Feladat 1
SELECT szh.SZALLAS_ID, szh.SZALLAS_NEV, COUNT(sz.KLIMAS)
FROM Szallashely szh JOIN Szoba sz ON szh.SZALLAS_ID=sz.SZALLAS_FK
WHERE sz.KLIMAS='i'
GROUP BY szh.SZALLAS_ID, szh.SZALLAS_NEV

```

```sql
--- Feladat 2
SELECT szh.SZALLAS_NEV,
sz.SZOBA_ID, 
sz.SZALLAS_FK, 
sz.SZOBA_SZAMA,
sz.FEROHELY, 
sz.POTAGY, 
sz.KLIMAS
FROM Szoba sz JOIN Szallashely szh ON szh.SZALLAS_ID=sz.SZALLAS_FK
LEFT JOIN Foglalas f ON f.SZOBA_FK=sz.SZALLAS_FK
```
