# **1.**

NULL értéket tartalmazó cellák száma a SZALL\_CIM oszlopban

NULL értéket tartalmazó cellák százalékos aránya

```sql
SELECT COUNT(*) AS 'Szállítási cím NULL értékek száma', 
        COUNT(*)*100.0/(SELECT COUNT(*) FROM Rendeles)
        AS 'Szállítási cím NULL értékek százaléka'
FROM Rendeles 
WHERE SZALL_CIM IS NULL
```

NULL értékű és nem NULL értékű cellák eloszlása kördiagramon ábrázolva

# 3.

Adott megrendelő <span style="color: var(--vscode-foreground);">rendeléseinek száma</span>

Adott megrendelő megrendeléseinek százalékos aránya

```sql
SELECT [LOGIN]
    ,COUNT(*) as 'Megrendelések száma'
    ,COUNT(*)*100.0/(SELECT COUNT(*) FROM Rendeles) as 'Megrendelések százalékos aránya'
FROM Rendeles
GROUP BY [LOGIN]
ORDER BY 3 DESC
```

Megrendelések száma megrendelőnként oszlopdiagramon megjelenítve
